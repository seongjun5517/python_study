### 딕셔너리(dict)

```python
'''
<딕셔너리(dict)>
 - 여러개의 값을 가지고 다닐 수 있는 타입 중의 하나
 - 1개의 데이터(값)는 key의 value라는 값을 쌍으로 가지고 있음
 - 1개의 데이터(값)에서 실제 값은 value에 저장되며,
   -> value에 저장되어 있는 실제 값을 추출하기 위해서는 key로 접근하여 추출합니다.
 - 딕셔너리(dict)를 표현하는 기호는 중괄호({})를 사용함
 - 리스트는 인덱스 번호를 이용하여 값에 접근하고,
   딕셔너리는 key값을 이용해서 값(value)에 접근함
   따라서, 딕셔너리느 인덱스 번호에 의미를 두지 않음(순서 무관)
 - 딕셔너리에서 key값은 고유한 값만 사용해야함(중복 허용하지 않음


<딕셔너리가 사용하는 key와 value에 저장할 수 있는 타입>
 - key : 정수, 실수 , 문자열 (주로 정수 또는 문자열을 사용)
 - value : 모든 타입 사용가능(정수, 실수, 문자열, bool, 리스트, 딕셔너리, 튜플 등등.... )
'''
```

```python
"""
dict2 딕셔너리 데이터를 이용해서
 - list1 및 list2 리스트 변수 생성
 - dict2의 key값 뒤에 숫자가 짝수이면 list1에 value값 추가
 - dict2의 key값 뒤의 숫자가 홀수이면 list2에 value값 추가
 - 딕셔너리 변수 dict3 생성해서 아래와 같이 저장 
   --> key값 : 짝수, 홀수
   --> value값 : list1, list2
"""
```

```python
### list1 및 list2 리스트변수 생성
list1 = []
list2 = []

### dict2 내에 key 값 뒤에 숫자가 짝수인지 홀수인지 체크

for k, v in dict2.items():
    # 짝수라면..
    if int(k[4:]) % 2 == 0 : 
        list1.append(v)
    # 홀수라면..
    else :
        list2.append(v)
### dict3 딕셔너리 변수에 넣기
dict3 = {"짝수" : list1, "홀수" : list2}
print(dict3)
```

```python
### dict3에 각 value값을 모두 각각 출력해 주세요.
# 출력예시
# 홍길동2
#  ...
# 홍길동100
# 홍길동1
#  ...
# 홍길동99

for list_v in dict3.values() :
    for v in list_v :
        print(v)
```
