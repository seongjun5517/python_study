### 딕셔너리(dict)

```python
'''
<딕셔너리(dict)>
 - 여러개의 값을 가지고 다닐 수 있는 타입 중의 하나
 - 1개의 데이터(값)는 key의 value라는 값을 쌍으로 가지고 있음
 - 1개의 데이터(값)에서 실제 값은 value에 저장되며,
   -> value에 저장되어 있는 실제 값을 추출하기 위해서는 key로 접근하여 추출합니다.
 - 딕셔너리(dict)를 표현하는 기호는 중괄호({})를 사용함
 - 리스트는 인덱스 번호를 이용하여 값에 접근하고,
   딕셔너리는 key값을 이용해서 값(value)에 접근함
   따라서, 딕셔너리느 인덱스 번호에 의미를 두지 않음(순서 무관)
 - 딕셔너리에서 key값은 고유한 값만 사용해야함(중복 허용하지 않음

<딕셔너리가 사용하는 key와 value에 저장할 수 있는 타입>
 - key : 정수, 실수 , 문자열 (주로 정수 또는 문자열을 사용)
 - value : 모든 타입 사용가능(정수, 실수, 문자열, bool, 리스트, 딕셔너리, 튜플 등등.... )
'''
```

```python
### 딕셔너리 생성하기

dict_1 = {"name":"홍길동", "age":"44","key":"value" }

dict_1
```

```python
### 값 추출하기 : 딕셔너리는 key로 접근해서 값을 추출합니다,
# - 딕셔너리에서 값을 추출하는 개념 : key를 줄테니, value를 달

dict_1["name"]
```

```python
### 값 수정하기 : 수정하고자 하는 key에 접근해서 값을 대입하면 됨
# - (중요) : 존재하지 않는 key를 이용해서 대입하게 되면 새로운 key:value가 생성됨

dict_1["name"] = "이순신"

### 기존에 key가 없으면 값(key:value)이 추가됨
# ***  key가 있으면 수정, 없으면 추가
dict_1["name2"] = "이순신2"
dict_1
```

```python
dict_2 = {1:"홍길동", 2:"이순신", 3:"김유신"}
dict_2[1]
```

```python
dict_3 = {0 : dict_1, 1 : dict_2}
dict_3
```

```python
### 일반적으로 사용되는 딕셔너리 형태 (1)
# - 행렬 데이터 생성시 도는 데이터 수집시 아래 형태로 제공을 많이 함

name_dict = {"name" : ["홍길동", "김유신", "이순신"]}
age_dict = {"age" : [40, 50, 60]}
area_dict = {"area" : ["서울", "부산", "대구"]}

list_dict = [name_dict, age_dict, area_dict]
list_dict
```

```python
### 일반적으로 사용되는 딕셔너리 형태 (2)
# - 행렬 데이터에서 조회 결과로 많이 출력함(Database에서 조회 결과는 아래와 같은 형태임)
row1_dict = {"name" : "홍길동", "age" : 40, "area" : "서울"}
row2_dict = {"name" : "김유신", "age" : 50, "area" : "부산"}
row3_dict = {"name" : "이순신", "age" : 60, "area" : "대구"}

rows_list = [row1_dict, row2_dict, row3_dict]
rows_list
```

```python
### key값들만 추출하기 : keys() 함수 사용
# - 반환결과는 list 타입 입니다.
# - 반복(for)문의 반복 패턴에 사용해서, 딕셔너리 내에 data를 추출할 수 있음
dict_1.keys() ,type(dict_1.keys())
```

```python
### value값들만 추출하기 : value()함수 사용
# - 반환결과는 리스트(list) 타입
# - values()만 단독으로 사용하는 경우는 거의 없음
dict_1.values()
```
