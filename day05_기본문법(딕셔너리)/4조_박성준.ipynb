```python
'''
<회원관리 프로그램>
 - 사용할 기능 : 신규 추가, 수정, 삭제, 검색(이름으로), 검색(전체목록), 종료
 - 사용할 데이터 형태 예시 : [{"name" : "홍길동", "age" : "55", "area" : "서울"}
                              {"name" : "이순신", "age" : "55", "area" : "서울"}
                              {"name" : "김유신", "age" : "55", "area" : "서울"}]
                            
 - 최초 메세지 값
   아래 번호 중에 하나를 선택해 주세요(번호로 이력하여 주세요):
       1. 리스트에 신규값 추가하기
       2. 수정하기
       3. 삭제하기
       4. 검색(이름으로 조회)
       5. 검색(전체 목록 조회)
       6. 종료

 - 새로운 주피터 파일 만들어서, 파일명은 "0조_개인이름.ipynb" 로 구글드라이브에 제출
   (제출위치 : 구글드라이브 > day05_기본문법(딕셔너리) > 회원관리_프로그램(개별제출))
   
 - 조건, 
   * 이름은 중복을 허용하지 않습니다.
   * 무한 반복 하면서 계속 수행
   * 다양한 예외를 고민하셔서 발생할 수 있는 처리 조치

'''
```

```python
dict_list = {}
dict_age
list1 = []
close = 0
while 1:

    if close == "종료":
        
        break
    
    chlist = input("아래 번호 중에 하나를 선택해주세요.1. 리스트에 신규값 추가하기 2. 수정하기 3. 삭제하기 4. 검색(이름으로 조회) 5. 검색(전체 목록 조회) 6. 종료")


    if chlist == "1" :
        
        # 리스트에 신규값 추가하기
        
        dict_list_copy = dict_list.copy()
        name = input("이름을 입력해주세요")
        age = input("나이를 입력해주세요")
        area = input("지역를 입력해주세요")
        
        dict_list_copy["name"] = name
        dict_list_copy["age"] = age
        dict_list_copy["area"] = area
        list1.append(dict_list_copy)
        list1
    if chlist == "2" :
        # 수정하기
        change_info = input("수정하실 정보를 입력(이름,나이,지역)")
        
        if change_info == "이름" :
            w_name = input("잘못 입력한 이름 입력")
            for i in range(0,len(list1),1):
                #print(list1[i]["name"])
                if list1[i]["name"] == w_name :
                    change_name = input("다시 입력할 이름")
                    list1[i]["name"] = change_name
                    print(f"{w_name}님에서 {change_name}님으로 변경되었습니다.")
                    break
        if change_info == "나이" :
            w_age = input("잘못 입력한 나이 입력")
            for i in range(0,len(list1),1):
                if list1[i]["age"] == w_age :
                    change_age = input("다시 입력할 나이")
                    list1[i]["age"] = change_age
                    print(f"{w_age}살에서 {change_age}살로 변경되었습니다.")
                    break
        
        if change_info == "지역" :
            w_area = input("잘못 입력한 지역 입력")
            for i in range(0,len(list1),1):
                if list1[i]["area"] == w_area :
                    change_area = input("다시 입력할 지역")
                    list1[i]["area"] = change_area
                    print(f"{w_area}지역에서 {change_area}지역으로 변경되었습니다.")
                    break
    if chlist == "3" :
        
        #삭제하기
        
        del_info = input("삭제할 정보의 이름을 입력해주세요")
        for i in range(0,len(list1),1):
            if list1[i]["name"] in del_info:
                del_info_chk = input(f"삭제할 정보가 {list1[i]['name']} 맞습니까? 맞으면 예를 입력해주세요")
                if del_info_chk == "예":
                    del list1[i]
                    print("삭제완료")
                    break
                    
            else :
                print("해당정보가 없습니다")
                break
    if chlist == "4" :
        # 검색(이름으로 조회)
        
        sh_name = input("조회하실 정보의 이름을 입력해주세요")
        
        for i in range(0,len(list1),1):
            if list1[i]["name"] == sh_name :
                print(f'{sh_name}의 정보 이름 :{list1[i]["name"]} 나이 : {list1[i]["age"]} 지역 : {list1[i]["area"]}')
            else :
                print("해당정보가 없습니다.")
                break
    if chlist == "5" :
        # 검색(전체목록 조회)
        for i in range(0,len(list1),1) :
            print(f"이름 : {list1[i]['name']} 나이 : {list1[i]['age']} 지역 : {list1[i]['area']}" )
    
    if chlist == "6" :
        close = "종료"
```
