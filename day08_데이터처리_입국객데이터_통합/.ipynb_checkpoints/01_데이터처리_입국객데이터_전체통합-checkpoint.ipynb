### 국적 데이터 처리하기(대륙이름이 포함되어 있음)

```python
### 국적 데이터의 고유한 값들만 추출하기
# - 고유한값 추출 함수(시리즈 클래스에서 제공하는 함수) : unique()
print(len(kto_201001["국적"].unique()))
kto_201001["국적"].unique()


### 대륙 이름 : '아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'
### 실제 이름 : '아시아', '아메리카', '유럽', '오세아니아', '아프리카', '기타대륙', '교포'
```

```python
### 국적 컬럼에서 대륙 데이터 제거하기
# 제거할 대륙 이름 : '아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'

chk_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']

# 필터링
chk = kto_201001["국적"].isin(chk_list)

# 대륙이 포함된 데이터 전체 조회
kto_201001[chk == True]

# 대륙이 포함되지 않은 데이터 전체 조회
kto_201001_country = kto_201001[chk == False]

# 현재까지 최종 처리된 데이터프레임
kto_201001_country


# 최종 처리 결과에서 고유한 국적 값만 추출해 보기 : 대륙이름 없는지 검증
kto_201001_country["국적"].unique()
```

```python
# 현재까지 최종 처리된 데이터프레임
kto_201001_country
```

```python
### "행 인덱스 값"을 순차적으로 재정의 하기
# - 데이터프레임 클래스에서 제공하는 reset_index() 함수 사용
# - drop=True 속성 : 기존 "행 인덱스값"은 컬럼으로 사용하지 않겠다는 의미
kto_201001_country_newindex = kto_201001_country.reset_index(drop=True)
kto_201001_country_newindex
```

### 대륙 컬럼 추가하기

```python
### 리스트 + 리스트 = 1개 리스트로 합쳐짐
["아시아"]*2 + ["아메리카"]*3
```

```python
### 실제 이름 : '아시아', '아메리카', '유럽', '오세아니아', '아프리카', '기타대륙', '교포'
# - 아시아 : 25개
# - 아메리카 : 5개
# - 유럽 : 23개
# - 오세아니아 : 3개
# - 아프리카 : 2개
# - 기타대륙 : 1개
# - 교포 : 1개

# print(25+5+23+3+2+1+1, len(kto_201001_country_newindex))
# kto_201001_country_newindex

### 대륙 컬럼에 사용할 열 데이터를 리스트로 만들기
# 리스트 60개 데이터
contients = ["아시아"]*25 + ["아메리카"]*5 + ["유럽"]*23 + ["오세아니아"]*3 + ["아프리카"]*2 + ["기타대륙"]*1 + ["교포"]*1
# print(len(contients), contients)

### 대륙 컬럼 추가하기
kto_201001_country_newindex["대륙"] = contients
kto_201001_country_newindex
```

### 자국 내에서 관광의 비율을 확인하기위한컬럼 추가하기

```python
### 컬럼명 : 자국내_관광객비율
# 자국내_관광객비율(%) = (자국내 관광객수 / 자국내 계) * 100


a = kto_201001_country_newindex["관광"]
b = kto_201001_country_newindex["계"]
kto_201001_country_newindex["자국내_관광객비율"] = round(a / b * 100, 3)


kto_201001_country_newindex
```

### 국가별 관광객 비율 컬럼 추가

```python
### 컬럼명 : 국가별_관광객비율
# 1. 국가 전체에 대한 관광객수 총합계가 필요함
total_sum = kto_201001_country_newindex["관광"].sum()
# print(total_sum)

# 2. 국가별_관광객비율(%) = (국가별 관광객수 / 충관광객수)* 100
kto_201001_country_newindex["국가별_관광객비율(%)"] = round((a / total_sum) * 100, 3)
kto_201001_country_newindex
```

### 샘플링 처리 내용 정리

```python
"""
<지금까지 데이터 처리 순서>
 1. 파일 읽어들이기
 2. 기준년월 컬럼 추가
 3. 국적 컬럼에서 대륙이름 제거
 4. 데이터프레임 인덱스 재정의
 5. 대륙 컬럼 추가
 6. 자국내 관광객 비율 컬럼 추가
 7. 국가별 관광객 비율 컬럼 추가


 <통합을 위한 처리 순서>
 - 조별로 논의하기...

 1. 여러 파일 읽어들이기 위한 "반복문"만들기
    >2. "각 파일명 " 자동 생성
    >3. 위 "샘플링 코드" 그대로 사용
    >4. 깨끗한 데이터프레임에 각 처리된 데이터프레임"행단위 추가"하기
"""
```

```python
### 기준년월 컬럼 추가
kto_201001["기준년월"] = "2010-01"

### 국적 컬럼에서 대륙 데이터 제거하기
chk_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
chk = kto_201001["국적"].isin(chk_list)
kto_201001[chk == True]
kto_201001_country = kto_201001[chk == False]

### 데이터프레임 인덱스 재정의
kto_201001_country_newindex = kto_201001_country.reset_index(drop=True)

### 대륙 컬럼 추가
contients = ["아시아"]*25 + ["아메리카"]*5 + ["유럽"]*23 + ["오세아니아"]*3 + ["아프리카"]*2 + ["기타대륙"]*1 + ["교포"]*1
kto_201001_country_newindex["대륙"] = contients

### 자국내 관광객 비율 컬럼 추가
a = kto_201001_country_newindex["관광"]
b = kto_201001_country_newindex["계"]
kto_201001_country_newindex["자국내_관광객비율"] = round(a / b * 100, 3)

###국가별 관광객 비율 컬럼추가
total_sum = kto_201001_country_newindex["관광"].sum()
kto_201001_country_newindex["국가별_관광객비율(%)"] = round((a / total_sum) * 100, 3)


kto_total = pd.concat([kto_total, kto_201001_country_newindex], ignore_index=True)
```
