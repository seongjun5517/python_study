```python
### 입국객수의 데이터 중에 500이상인 데이터만 조회해 주세요~
# 1. 입국객수의 데이터 열단위 조회
# 2. 조회된 데이터를 기준으로 조건처림

a = sample_1["입국객수"] >= 500


temp_df = sample_1[a == True]
temp_df

### 행과 열 인덱스의 형태***
# - 눈에 보이는 행과 열의 인덱스는 메모리에 저장된 값("인덱스 값")을 의미하며
# - 컴퓨터 메모리에는 실제 행과 열의 "인덱스 번호"를 순차적으로 가지고 있습니다.
# - 행렬 데이터 프레임은 행과 열의 인덱스를 "값"과 "번호"로 관리하며,
#   --> 눈에 보이는 건 "값", 실제 메모리는 순차적인"번호"로 관리 됨
```

```python
### 국적코드의 값이 A01이고,
# 입국객수의 값이 150000 이상인,
# 국적코드와 입국객수의 데이터 조회하기

### 조건이 2개 이상 : 논리연산자도 사용해야 합니다.
# - 논리 연산자의 왼쪽/오른쪽 조건들은 소괄호로 묶어야 오류가 안납니다.

# 1. 조건1처리
c1 = sample_1["국적코드"] == "A01"

# 2. 조건2 처리
c2 = sample_1["입국객수"] >= 150000

# 3. 조건1과 조건2를모두 만족하는 전체 데이터 조회
c3 = c1 & c2
temp_df = sample_1[c3 == True]

# <열 단위 조회>
# 4. 전체 데이터에서 필요한 컴럼만 조회
temp_df[["국적코드", "입국객수"]]



```

```python
### 위 처리 순서 1~3를 한줄로 합치기

temp_df = sample_1[(sample_1["국적코드"] == "A01") &
         (sample_1["입국객수"] >= 150000) == True][["국적코드", "입국객수"]]
temp_df
```

```python
### 국적코드의 값이 A01 또는 A18인
#   국적코드, 입국객수의 데이터 조회하기

n1 = (sample_1["국적코드"] == "A01" ) | (sample_1["국적코드"] == "A18")
 
sample_1[n1][["국적코드","입국객수"]]
```

```python
### or 조건인 경우 함수를 이용하여 처리 가능
# - 포함관계의 개념을 적용한 함수 사용
# - 파이썬 기본 문법에서 리스트에 포함되어 있는지 확인하는 in 과 유사한 개념이 존재함
# - isin([])함수 사용
c = sample_1["국적코드"].isin(["A01","A18"])
sample_1[c][["국적코드", "입국객수"]]
```

```python
### sample_1을 기준으로 code_code_master의 국정명 컬럼 합치기
# - 합치기 : 의미적을 포함시키기가 맞을 듯....
# - 함수 : 판다스에서 제공하는 merge() 사용 (데이터프레임에서 제공하지 않음)
# - 오른쪽 데이터의 컬럼들 중에 매핑에 사용되지 않은 컬럼들이 왼쪽으로 추가됩니다.

# - left : 원본 데이터프레임 변수명
# - right : 포함할 대상 데이터프레임 변수명

# - how : 어떻게 매핑 할 것인지 정의
#      -> inner : 왼쪽과 오른쪽의 매핑 컬럼의 값이 같은 행에 대해서만 처리함
#      -> left : 왼쪽 데이터는 모두 조회시키되, 오른쪽의 매핑값이 있으면 가지고 오고,
#                                                                없으면 비워 둡니다.
#      -> right : 오른쪽 데이터는 모두 조회시키되, 왼쪽의 매핑값이 있으면 가지고 오고,
#                                                                없으면 비워 둡니다.
#           --> ** 보통 왼쪽 기준으로 작성합니다.

# - left_on : 원본에서 매핑에 사용할 컬럼명
# - right_on : 대상데이터에서 매핑에 사용할 컬럼명
# - 매핑에 사용할 데이터의 성격은 같고 컬럼명이 틀리더라도, 
#    -> 컬럼명은 그대로 사용하면 됩니다.

pd.merge(left = sample_1, 
         right = code_master,
         how = "inner",
         left_on = "국적코드",
         right_on = "국적코드")
```

```python
temp_df = pd.merge(left = sample_1, 
                     right = code_master,
                     how = "left",
                     left_on = "국적코드",
                     right_on = "국적코드")
temp_df.info()
```

```python
### 우리가 사용할 데이터 "Left" 사용
sample_1_code = pd.merge(left = sample_1, 
                     right = code_master,
                     how = "left",
                     left_on = "국적코드",
                     right_on = "국적코드")
sample_1_code
```

### 데이터 통합 : 행 단위 통합하기

```python
### 같은 성격의 데이터 + 열의 갯수 동일해야함
# 현재 사용할 마지막 처리된 변수명


sample_1_code
```

```python
### 통합하기
# 1. 통합에 사용할 데이터 파일 : sample_2.xlsx
# 2. 통합에 사용할 파일 읽어 들이기
#    - 데이터프레임 변수명 : sample_2

# 3. 원본(sample_1_code) 데이터프레임의 컬럼들과 동일하게 데이터 처리
#    - 최종 데이터프레임 변수명 : sample_2_code

# 1.
file_path = "./files/sample_2.xlsx"
# 2.
sample_2 = pd.read_excel(file_path,
                         header=1,
                         usecols="A:C",
                         skipfooter=2)

sample_2["기준년월"] = "2019-12"

# 3.
sample_2_code = pd.merge(left = sample_2,
                         right = code_master, 
                         how = "left",
                         left_on = "국적코드",
                         right_on = "국적코드")

sample_2_code
```

```python
sample_1_code
```

```python
### 최종 합치기 전 데이터프레임 변수명
sample_1_code
sample_2_code
```

```python
### 해단위(밑으로) 합치기
# sample_1_code 데이터를 시작을 밑으로
# sample_2_code 데이터를 추가하기
# - 판다스에서 제공하는 concat([데이터프레임1, 데이터프레임2,.....데이터프레임n])함수를 사용
# - ignore_index=True : 데이터를 행단위로 통합시 "인덱스 값" 을 순차적으로 만들기 위한 속성
#                     : 속성을 생략하면 순차적이지 않고, 원본의 "인덱스 값"이 그대로 사용됨
sample = pd.concat([sample_1_code, sample_2_code], ignore_index=True)
sample
```

### 파일로 저장하기


```python
### 데이터프레임 클래스에서 저장하는 멤버함수를 제공합니다.
# - to_xxx()함수 사용

# - 저장할 위치 지정
save_path = "./files/sample.xlsx"

# - 저장하기
# - index=False : 파일 저장시 데이터프레임의 "행 인덱스 값"은 제외 시키기
sample.to_excel(save_path, index=False)
```
