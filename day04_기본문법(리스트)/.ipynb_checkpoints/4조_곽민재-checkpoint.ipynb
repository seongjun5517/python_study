```python
members = []
age = []
address = []

while True:
    # 번호를 선택하기 위해서 먼저 아래처럼 print함수를 이용해서 나타낸다
    print("아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요) :")
    print("1. 리스트에 신규값 추가하기")
    print("2. 수정하기")
    print("3. 삭제하기")
    print("4. 검색(이름으로 조회)")
    print("5. 검색(전체 목록 조회)")
    print("6. 종료")
    
    # 선택할 번호를 선택하기 위해서 input를 이용한다.
    menu = input("선택해주세요 => ")

    # 1. 리스트에 신규 추가하기
    
        # if문으로 1번을 선택한다면 변수 a를 이용해서 이름을 검색할 input문을 사용합니다. 그리고 이름, 나이, 주소 각각에 맞는 input를 사용하고 
        
    if menu == '1' :
        a = input("이름 : ")
        # 이름은 중복되면 안되기 때문에 이름 중복에 관련된 if문을 만듭니다.
        if a in members :
            print("중복된 이름입니다")
            break
            
        b = input('나이 : ')
        c = input('주소 : ')
        
        # 그리고 각각의 list문에 항목을 추가하기 위해서 append문을 사용합니다. 
        
        members.append(a)
        age.append(b)
        address.append(c)
        print(f"{members} 회원이 추가되었습니다.")
        
    # 2. 수정하기
        # if문으로 2번을 선택한다면 변수 a를 이용해서 회원이름을 수정 할 input문을 사용합니다.
        # 수정할 회원의 이름을 변수와 함께 input를 사용하고 수정할 이름의 인덱스 위치를 알기 위해서 index문을 사용합니다. 
    
    elif menu == '2' :
        a = input("수정할 회원 이름 : ")
        b = members.index(a)
        
    # 이름, 나이, 주소를 바꾸기 위해서 list[b] = input("")형태로 입력합니다.
        members[b] = input('변경할 이름을 입력해주세요 :')
        age[b] = input('변경할 나이를 입력해주세요 :')
        address[b] = input('변경할 주소를 입력해주세요 :')
    
        print(f"{members} 회원 정보가 수정되었습니다.")

     # 3. 삭제하기
        
        # if문으로 3번을 선택한다면 변수 a를 이용해서 회원 이름을 삭제할 input문을 사용합니다.
        # 삭제할 회원의 이름을 변수와 함께 input를 사용하고 삭제할 이름의 인덱스 위치를 알기 위해서 index문을 사용합니다. 
   elif menu == '3' :
        print(members)
        a = input("삭제할 회원 이름")
        b = members.index(a)
        # 그리고 해당 회원이 존재하지 않을 수 있기 때문에 해당 회원이 없을 때의 if문을 사용합니다.
        if a not in members :
            print("해당 회원이 존재하지 않습니다.")
            
    # del문을 사용해서 이름에 해당하는 나이와 주소도 삭제합니다.
        else :
            del members[b]
            del age[b]
            del address[b]
        
    # 4. 검색(이름으로 조회)
    #  if문으로 4번을 선택한다면 변수 a를 이용해서 회원이름을 검색할 input문을 사용합니다.

    elif menu == '4' :
        a = input("검색할 회원 이름: ")
        b = members.index(a)
    # 검색할 회원 이름이 없으면 오류가 나기 때문에 오류가 안 생기고자 회원이름이 없을 때의 if문을 만듭니다. 
        if a not in members:
            print("검색 결과가 없습니다.")
            continue
     # 회원이름을 검색하면 검색할 회원에 맞는 이름, 주소, 나이를 다 표시하게 print문을 사용합니다.   
        else :
            print(f"이름 : {members[b]}, 나이: {int(age[b])}, 주소: {address[b]}")
        
   # 5. 검색(전체 목록 조회) 
    
    #  if문으로 5번을 선택한다면 전체 목록이 다 나타나게 print문을 사용합니다. 
    elif menu == '5' :
         if not members:
             print("등록된 회원이 없습니다.")
         else :
            print(f"이름 : {members}, 나이 : {age}, 주소 : {address}")

    # 6. 종료
     #  if문으로 6번을 선택한다면 프로그램이 종료된다는 print문을 만듭니다.

    elif menu == '6':
        print("프로그램을 종료합니다.")
        break

    # 그리고 1 ~ 6 번을 선택하지 않았다면 오류가 생기지 않게 이와 관련된 print문도 만듭니다. 
    else :
        print("1 ~ 6번 중에서 선택해 주세요.")

    
```
