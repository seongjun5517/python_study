```python
'''
<회원관리 프로그램>
 - 사용할 기능 : 신규 추가, 수정, 삭제, 검색(이름으로), 검색(전체목록), 종료
 - 사용할 데이터 : 이름, 주소, 나이
 - 최초 메세지 값
   아래 번호 중에 하나를 선택해 주세요(번호로 이력하여 주세요):
       1. 리스트에 신규값 추가하기
       2. 수정하기
       3. 삭제하기
       4. 검색(이름으로 조회)
       5. 검색(전체 목록 조회)
       6. 종료
 - 조건, 
   * 이름은 중복을 허용하지 않습니다.
   * 무한 반복 하면서 계속 수행
   * 다양한 예외를 고민하셔서 발생할 수 있는 처리 조치

'''
```

```python
# 빈 리스트를 만든다
list_name = []
list_addr = []
list_age = []
close = 0
# 전체 반복
while 1: 
    # 종료를 입력 받으면 break
    if close == "종료" :
        # 전체 종료한다.
        break
    # chlist 안에 있는 번호를 입력한다
    chlist = input("아래 번호 중에 하나를 선택해주세요.1. 리스트에 신규값 추가하기 2. 수정하기 3. 삭제하기 4. 검색(이름으로 조회) 5. 검색(전체 목록 조회) 6. 종료")
    # 만약 입력받은 값이 1이라면
    if chlist == "1" :   
        # 이름을 입력받는다
        name = input("이름을 입력해주세요")  

        # name리스트안에 입력받은 이름이 없으면 주소, 나이를 입력 받는다.
        if name not in list_name :
            # name 리스트에 이름을 넣는다
            list_name.append(name)
            # 주소를 입력받는다
            addr = input("주소을 입력해주세요")
            # addr 리스트에 주소를 넣는다
            list_addr.append(addr)
            # 나이를 입력받는다
            age = input("나이을 입력해주세요")
            # age 리스트에 나이를 넣는다
            list_age.append(age)
        # 중복된 이름이면 다시 입력한다.
        else:
            print("중복된 이름입니다 다시 입력해주세요")

    # 입력받은 값이 2라면
    if chlist == "2" :

        
        change = ["이름","주소","나이"]
        
        # 수정할 정보 "이름", "주소", "나이"를 입력 받으면 while문을 종료한다
        while 1 :
            change_infor = input("수정할 정보를 입력해주세요(이름, 주소, 나이)")
            if change_infor == "이름" :
                break 
            elif change_infor == "주소" :
                break 
            elif change_infor == "나이" : 
                break 
            # 그게 아니라면 다시 입력받는다
            if change_infor not in change :
                print("올바른 입력이 아닙니다. 다시 입력해주세요.")

        # "이름"을 입력받으면 
        if change_infor == "이름":
            # 수정하고싶은 이름을 입력받는다
            change_name = input("수정하고싶은 이름을 입력해주세요.")
            for i in range(0,len(list_name),1):
                if change_name == list_name[i]:
                    # 무엇으로 바꿀껀지 입력받는다
                    change_name1 =input("무엇으로 변경해드릴까요?")
                    list_name[i] =change_name1
                    print(f"{list_name[i]}로 변경되었습니다")
                    # 변경 되었으므로 멈춘다
                    break
                # 아니라면 다시 시작한다
                else :
                    print("이름이 틀립니다. 다시 시작해주세요")
        # 주소를 입력 받으면
        if change_infor == "주소":
            # 수정하고 싶은 주소를 입력받는다
            change_addr = input("수정하고싶은 주소을 입력해주세요.")
            for i in range(0,len(list_addr),1):
                if change_addr == list_addr[i]:
                    # 무엇으로 바꿀껀지 입력받는다
                    change_addr1 =input("무엇으로 변경해드릴까요?")
                    list_addr[i] =change_addr1
                    print(f"{list_addr[i]}(으)로 변경되었습니다")
                    # 변경 되었으므로 멈춘다
                    break
                # 입력받은 주소가 틀리면 다시 시작한다.
                else :
                    print("주소가 틀립니다. 다시 시작해주세요.")
        # 나이를 입력 받으면
        if change_infor == "나이":
            # 수정하고 싶은 나이를 입력받는다.
            change_age = input("수정하고싶은 나이을 입력해주세요.")
            for i in range(0,len(list_age),1):
                if change_age == list_age[i]:
                    # 무엇으로 바꿀껀지 입력받는다.
                    change_age1 =input("무엇으로 변경해드릴까요?")
                    list_age[i] =change_age1
                    print(f"{list_age[i]}세로 변경되었습니다")
                    # 변경 되었으므로 멈춘다
                    break
                # 입력받은 나이가 틀리면 다시 시작한다.
                else :
                    print("나이가 틀립니다. 다시 시작해주세요.")

    # 3을 입력받으면
    if chlist == "3" :
        # 삭제할 정보를 입력 받는다
        remover = input("삭제할 정보를 입력해주세요")
        remove1 = list_name.index(remover)
        # 삭제할 정보가 맞는지 확인 받는다
        agree = input(f"{list_name[remove1]}님의 정보 이름:{list_name[remove1]}, 주소:{list_addr[remove1]}, 나이:{list_age[remove1]}를 정말 삭제하시겠습니까? 맞으시면 '예' 아니면 '아니요'를 쳐주세요")
        if agree == "예" :
            for i in range(0, len(list_name), 1):
                if remover == list_name[i]:
                    del list_name[remove1]
                    print("처리 완료")
    # 4를 입력받으면
    if chlist == "4" :
    
        # 이름을 입력 받는다.
        sh_name = input("조회하실 이름을 입력해주세요")
        # 입력 받은 이름으로 인덱스번호를 확인 후 정보를 확인한다.
        sh_name2 = list_name.index(sh_name)
        print(f"{list_name[sh_name2]}님의 정보 이름:{list_name[sh_name2]} 주소:{list_addr[sh_name2]} 나이:{list_age[sh_name2]}")
    # 5를 입력 받으면
    if chlist == "5" :
        
        # 이름의 인덱스 길이 만큼 반복하고 정보를 출력한다.
        for i in range(0, len(list_name), 1) :
            print(f"{list_name[i]}님의 정보 이름:{list_name[i]} 주소:{list_addr[i]} 나이:{list_age[i]}")

    # 6을 입력받으면
    if chlist == "6" :
        # close 에 종료를 넣는다
        close = "종료"
    
```
