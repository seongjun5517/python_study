```python
"""
<회원관리 프로그램>
 - 사용할 기능 : 신규 추가, 수정, 삭제, 검색(이름으로,) 검색(전체 목록), 종료
 - 사용할 데이터 : 이름, 나이 주소
 - 최초 메시지 값
     아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요):
         1. 리스트에 신규값 추가하가
         2. 수정하기 /이름은 수정하기 불가능
         3. 삭제하기 / 정말 삭제하시겠습니까? 한번 더 물어봐라
         4. 검색(이름으로 조회)
         5. 검색(전체 목록 조회)
         6. 종료
 - 조건, 
     * 이름은 중복을 허용하지 않습니다.
     * 무한 반복하면서 계속 수행
     * 다양한 예외를 고민하셔서 발생할 수 있는 처리 조치

 - 새로운 주피터 파일을 만들어서, 파일명은 "0조_개인이름.ipynb"로 구글 드라이브에 제출
"""
```

```python
name_list = []
age_list = []
addr_list= []



# 무한 반복
while 1 :
    # 최초 메시지 값 
    print("""아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요)\n1. 리스트에 신규값 추가하기\n2. 수정하기\n3. 삭제하기\n4. 검색(이름으로 조회)\n5. 검색(전체 목록 조회)\n6. 종료""")

    # input값 받기
    f_an = input("입력값 :")
    print("=============================================")


    
    # if f_an == 1
    # 리스트에 신규값 추가하기
    if f_an == "1":
        name = input("이름을 입력해 주세요 :")
        # 이름은 중복을 허용하지 않는다.
        if name in name_list:
            print(f"이미 [{name}]님은 가입된 이력이 있습니다.")
            print("=============================================")
            continue
            
    
        while 1:
            age = input("나이를 '숫자'로 입력해 주세요(초기화면으로 돌아가려먼 x를 입력해주세요) :")
            if age == "x":
                print("초기화면으로 돌아갑니다.")
                print("=============================================")
                break
            else:
                #나이를 숫자로 "쉰"과 같이 글자로 입력하는걸 방지
                try:
                    age = int(age) # 문자는 정수화 하면 오류가 나기때문에 except문으로
                    addr = input("주소를 입력해 주세요 :")
                    name_list.append(name) # 모두 성공적으로 실행된 다음 list에 추가
                    age_list.append(age)
                    addr_list.append(addr)
                    print(f"[{name}]회원님이 신규 회원으로 등록되었습니다.")
                    print("=============================================")
                    break 
                except:
                    print("**숫자가 아닙니다.**")
                                
        
    
    
    # if f_an == 2
    # 리스트 수정하기
    elif f_an == "2":
        name = input("수정하실 분의 이름을 입력해주세요.")
    
        # 이름 검색
        if name in name_list: 
            idx = name_list.index(name)    
            # 나이 숫자로 받기
            while 1:
                age = input("수정하실 나이를 입력하세요.(처음으로 돌아가려먼 x를 입력해주세요) :")
                # 무한반복문은 빠져나갈 방법을 만들어 두기
                if age == "x":
                    print("초기화면으로 돌아갑니다.")
                    print("=============================================")
                    break
                else:
                    try:
                        age = int(age)
                        addr = input("수정하실 주소를 입력하세요:")
                        age_list[idx] = age
                        addr_list[idx] = addr
                        print(f"[{name}]회원님의 정보가 수정되었습니다.")
                        print("=============================================")
                        break 
                    except:
                        print("**숫자가 아닙니다.**")
                    

        
        else : 
            print(f"입력하신 [{name}]님은 가입된 이력이 없습니다.")
            print("=============================================")





    
    # if f_an == 3
    # 리스트 삭제하기 + 정말 삭제할지 한번 더 물어보기
    elif f_an == "3":
        name = input("삭제하실 분의 이름을 입력해주세요.")
        
        # 이름 검색
        
        if name in name_list: 
            idx = name_list.index(name)    
            print(f"[{name}]님의 가입이력이 존재합니다. 삭제하시겠습니까?")    
            while 1:
                s_an = input("네/아니오로 입력해주세요:")
                if s_an == "네":
                    del name_list[idx]
                    del age_list[idx]
                    del addr_list[idx]
                    print(f"[{name}]님의 가입이력이 삭제되었습니다.") 
                    print("=============================================")
                    break
                    
                elif s_an == "아니요":
                    print("삭제를 취소하셨습니다.")
                    break
        else : 
            print(f"입력하신 [{name}]님은 가입된 이력이 없습니다.")
            print("=============================================")



    
    # if f_an == 4
    # 이름으로 검색하기
    elif f_an == "4":
        name = input("검색하고자 하는 회원님의 이름을 입력해주세요.")

        # 이름 검색
        
        if name in name_list:
            idx = name_list.index(name) # 이름으로 인덱스 번호 찾기 
            print(f"[{name_list[idx]}]님 나이는[{age_list[idx]}]이고 주소는[{addr_list[idx]}]입니다")
            print("=============================================")
        else:
            print(f"입력하신 [{name}]님은 가입된 이력이 없습니다.")
            print("=============================================")
            



    
    # if f_an == 5
    # 전체 목록 조회 + 문장만들어서 전체 정보 다 보여줘
    elif f_an == "5":
        print("----- 전체 회원 목록 조회 -----")
        for i in range(len(name_list)):
            print(f"이름:{name_list[i]}    나이:{age_list[i]}    주소:{addr_list[i]}")
        print("=============================================")



    
    # if f_an == 6
    # 종료
    elif f_an == "6":
        print("종료하겠습니다.")
        break
    
    else:
        # 1~6 이외에 입력이 들어올 수 도 있음.
        print("다시 선택해주세요.")
        print("=============================================")
        
```
