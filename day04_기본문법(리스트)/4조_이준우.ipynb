```python
##### < 회원관리 프로그램 >
# 사용할 기능 : 신규 추가 , 수정, 삭제, 검색(이름으로), 검색(전체목록), 종료
# 사용할 데이터 : 이름, 나이, 주소
# 최초 메세지 값
# 아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요) :
    # 1. 리스트에 신규값 추가하기
    # 2. 수정하기
    # 3. 삭제하기
    # 4. 검색(이름으로 조회)
    # 5. 검색(전체 목록 조회)
    # 6. 종료
# - 조건
    # 이름은 중복을 허용하지 않습니다.
    # 무한 반복하면서 계속 수행
    # 다양한 예외를 고민하셔서 발생할 수 있는 처리 조치


# 예외 처리 오류 방지용
err_age = range(150)
# err_name = []
# err_addr = ['부산', '서울', '대구', '울산', '충청', '대전']

# 데이터관리용
data_name = []
data_age = []
data_addr = []

while 1:
# 메뉴 보이기
    print('아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요)')
    print('----------------------------')
    print('1. 리스트에 신규값 추가하기')
    print('2. 수정하기')
    print('3. 삭제하기')
    print('4. 검색(이름으로 조회)')
    print('5. 검색(전체 목록 조회)')
    print('6. 종료')
    print('----------------------------')
    # 메뉴 값 받기 1~6 받을예정
    cnt = input()
    
    # 1. 리스트에 신규값 추가하기(1번 선택)
# 메뉴값 1이면 이름 , 나이 , 주소 데이터 받기
# 1일 경우 a / 이름 데이터 입력받기
    if cnt == '1' :
        print('----------------------------')
        print('1. 리스트에 신규값 추가하기를 선택하셨습니다.')
        print('리스트에 신규값을 입력해 주세요.')
        
        a = input('1. 이름을 입력해주세요 :')
        if a in data_name:
            print('중복된 이름은 허용하지 않습니다. 다른 이름을 작성해주세요.')
            print('초기화면으로 돌아갑니다.')
            continue
# 2일 경우 b / 나이 데이터 입력받기                
        b = int(input('2. 나이를 입력해주세요 :'))
        if b not in err_age:
            print('정확한 나이를 입력해주세요. 잘못입력하셨습니다.')
            break
# 3일 경우 c / 주소 데이터 입력받기               
        c = input('3. 주소(지역)을(를) 입력해주세요 :')
# 크로스 체킹
        print("입력하신 데이터는")
        print(f'[ 이름 : {a} ], [ 나이 : {b} ], [ 지역 : {c} ]')
        print('-------- 맞으면 1번, 틀리면 2번을 입력해주세요 --------')
       
        cross_check = input()
# 크로스 체킹 1이면 그대로 등록        
        if cross_check == '1':
            data_name.append(a)
            data_age.append(b)
            data_addr.append(c)
            print(f'[ {a} ], [ {b} ], [ {c} ] 등록되었습니다.')
# 크로스 체킹 2이면 초기화면으로 돌아감            
        elif cross_check == '2':
            print('초기화면으로 돌아갑니다.')
# 이상한 번호 입력하면 초기화면으로 돌아가면서 오류문구 등록    
        else:
            print('정확한 번호를 입력해주세요')

    # 2. 수정하기 (2번선택)
# a 변수에 데이터 입력받아서 수정할 데이터받기    
    elif cnt == '2' :
        print('2. 수정하기를 선택하셨습니다.')
        print('수정하실 이름으로 조회해주세요.')
        print(data_name)
        a = input()
# 만약 리스트에 없는 데이터라면 처음화면으로 돌아감        
        if a not in data_name :
            print(f'{a}은(는) 데이터에 속해있지 않습니다.')
            print('처음 화면으로 돌아갑니다 ..... 3')
            print('처음 화면으로 돌아갑니다 ..... 2')
            print('처음 화면으로 돌아갑니다 ..... 1')
            print('----------------------------')
            print(f'')
# 리스트에 있다면 , 리스트 위치를 b 변수에 넣어서 동일 위치에 있는 이름 , 나이 , 주소를 출력         
        else:
            b = data_name.index(a)
            print(f'입력하신 내용은 이름 : [ {a} ], 나이 : [ {data_age[b]} ], 주소 : [ {data_addr[b]} ] 입니다.')
            print('이름을 바꾸려면 [ 1 ]번')
            print('나이를 바꾸려면 [ 2 ]번')
            print('주소를 바꾸려면 [ 3 ]번을 입력해주세요.')

# 이름 1번, 나이 2번, 주소 3번으로 원하는 값 변경을 요청 및 변환
            c = input()
            if c == '1':
                print('이름을 선택하셨습니다. 바꿀 이름을 입력해주세요.')
                d = input()
                print(f'기존 [ {a} ]에서 [ {d} ](으)로변경 되었습니다.')
                data_name[b] = d
            elif c == '2':
                print('나이를 선택하셨습니다. 바꿀 나이를 입력해주세요.')
                d = input()
                print(f'기존 [ {data_age[b]} ]살에서 [ {d} ]살로변경 되었습니다.')
                data_age[b] = d
            elif c =='3':
                print('주소를 선택하셨습니다. 바꿀 주소를 입력해주세요.')
                d = input()
                print(f'기존 [ {data_addr[b]} ]에서 [ {d} ](으)로변경 되었습니다.')
                data_addr[b] = d
            else:
                print('잘못 입력하셨습니다. 1번 2번 3번 중에 골라주세요 ')
# 3. 삭제하기
    elif cnt == '3' :
        print('3. 삭제하기를 선택하셨습니다.')
        print('삭제하실 이름을 입력해주세요.')
        print(data_name)
        a = input()
        
        if a not in data_name :
            print(f'{a}은(는) 데이터에 속해있지 않습니다.')
            print('처음 화면으로 돌아갑니다 ..... 3')
            print('처음 화면으로 돌아갑니다 ..... 2')
            print('처음 화면으로 돌아갑니다 ..... 1')
            print('----------------------------')
        else:
            b = data_name.index(a)
            del(data_name[b])
            del(data_age[b])
            del(data_addr[b])
            print(f'{a} 이름으로 된 리스트위치는 삭제되었습니다.')
            
# 4. 검색(이름으로 조회)
    elif cnt == '4' :
        print('4. 검색(이름으로 조회)을 선택하셨습니다.')
        print(data_name)
        a = input('검색할 이름을 입력해주세요')
        print('------------- 시 작---------------')
        print(' ')
        
# 리스트에 입력값이 있다면 위치에 관련된 내용을 반환
        if a in data_name :
            b = data_name.index(a)
            print(f'[ {a} ]님과 관련된 정보입니다.')
            print(f'[ {a} ]님은 [ {data_age[b]} ]살이고 [ {data_addr[b]} ]에 거주하고 있습니다. ')
            print(' ')
            print('------------- 끝 ---------------')
# 리스트에 입력값이 없다면 없는 데이터라고 출력
        else :
            print(f'[ {a} ]는 리스트에 없는 데이터입니다.')
        
# # 5. 검색(전체목록 조회)
    elif cnt == '5' :
        print('5. 검색(전체 목록 조회)을 선택하셨습니다.')
        print('------------- 시 작---------------')
        print(' ')
# for문을 이용하여 이름의 갯수만큼 있는 모든 데이터를 출력        
        for i in range(len(data_name)) :
            print(f'No.{i+1} [ {data_name[i]} ]님은 [ {data_age[i]} ]살이고 [ {data_addr[i]} ]에 거주하고 있습니다. ')
        
        print(' ')
        print('------------- 끝 ---------------')
# 6. 종료    
    elif cnt == '6' :
        print('6. 종료를 선택하셨습니다.')
        print('시스템을 종료합니다 ..... 3')
        print('시스템을 종료합니다 ..... 2')
        print('시스템을 종료합니다 ..... 1')
        print('----------------------------')
        break
# 7. 1~6이 아닌 다른 입력값 감지        
    else :
        print(f'정확한 번호를 입력해주세요. [ {cnt} ]은(는) 없는 번호입니다.')
        print('----------------------------')
```
