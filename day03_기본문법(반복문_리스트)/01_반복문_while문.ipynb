```python
'''

<while문>
 - 조건이 있는 경우와 없는 경우로 나누어서 반복을 수행 할 수 있음
 - for문의 경우에는 숫자의 범위 패턴을 이용해서 반복했다면,
  -> while문은 숫자의 범위 보다는, "조건"을 만족할 때까지 반복을 수행합니다.
 - while문은 가급적 종료하는 방법을 작성해야 합니다.
  -> 종료는 조건과 break 명령을 통해서 주로 처리됩니다.
 - 종료하는 방법을 작성하지 않으면 -> 무한반복(무한루프) 됩니다.
 - for문으로 작성한 것들은, 일반적으로 while문으로도 착성 가능합니다.

'''
```

```python
### 1부터 10까지의 총합을 출력하기
# - 우선 for문으로 적성해 주세요
result = 0
for i in range (1,11,1) :
    result = i + result
print(result)
```

```python
### 1부터 10까지의 총합을 while문으로 처리하기
# - 시작값 : 1
s_no = 0
# - 끝값 : 10
e_no = 10
# - 증가값 : 1

total  = 0
#while문 작성
# 종료하는 조건과 방법
while s_no <= e_no :
    total = total + s_no
    s_no = s_no + 1
print(total)

```

```python
### 1부터 10까지 반복하면서, 짝수인 값들만 출력해주세요.

#for문

for i in range(1,11,1):
    if i % 2 == 0 :
        print(i)
        
#while문

a = 1
b = 10
while a <= b :
    if a % 2 == 0 :
        print(a)
    a = a + 1
```

```python
### 1부터 10까지 반복하면서, 
# - 짝수값이 낭는 시점에
# - 반복을 강제 종료(break) 시키기

a = 1
b = 10
while a <= b :
    if a % 2 == 0 :
        print(a)
        break
    a = a + 1
```

```python
### 무한반복 시키기
# - 반복 조건을 True 또는 1로 정의하면 됩니다.
# - while True
#   while 1 (보통 1 사용)
a = 0
b = 10
while 1 :
    a = a + 1
    print(f"무한 반복 중...[{a}]")
    ### 무한반복의 경우 종료 조건을 무조건 만들고 시작해야 합니다.
    if a == b :
        print("무한반복을 종료합니다.")
        break
```

```python
### 무한반복 수행 
# - 반복할 때 마다 -> 임의 값 입력 받기 : x 입력 받기
# - 입력 받은 문자값이 x이면 반복을 강제 종료 시키기
# - x가 아니면, 1부터 1씩 증가된 값을 반복하면서 증가시키기

cnt = 0
while 1 :
    chk = input("임의 한개 문자를 입력해 주세요(x를 입력하면 종료합니다.)")    

    # 입력받은 값이 x인 경우에 강제 종료 시키기
    if chk == "x" :
        print("while문을 종료 합니다.")
        break
    cnt = cnt + 1
    print(f"count = {cnt}")
```

```python
### 구구단 출력하기 : while문으로 
# 2~9단까지 모두 출력
# 출력 예시
# ---------[2]단 ---------
# 2 x 1 = 2
# .....
# --------[9]단-----------
# 9 x 1 = 9

dan = 2
b = 9

while dan <= b :
    print(f"-----[{dan}]단-----")
    a = 1
    while a <= b :
        print(f"{dan} x {a} = {dan*a}")
        a = a + 1
    dan = dan + 1
        
# for i in range(1, 9 ,1) :
#     i = i + 1
#     print(f"-----[{i}]단-----")
#     for j in range (1, 10, 1):
#         d = i * j 
#         print(f"{i} x {j} = {d}")
    
```
