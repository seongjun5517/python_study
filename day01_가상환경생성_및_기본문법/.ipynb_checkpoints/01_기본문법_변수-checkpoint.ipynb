# 변수

```python
###변수 : 
# - 메모리에 저장된 값을 의미합니다.

a = 5 
# a라는 이름이 바라보는 메모리 주소공간에 5라는 값을 저장 해 둡니다.

b = 6 
# b라는 이름이 바라보는 메모리 주소공간에 6라는 값을 저장 해 둡니다.

c = a + b
# a,b라는이름이 바라보는 메모리의 값을 가지고 와서 두개의 값을 더하기 처리 (산술연산)이후에
# 값을 c라는 이름이 바라보는 메모리 주소공간에 11이라는 갑슬 저장해 둡니다.

# c라는 이름이 바라보는 메모리 주소공간에 a + b라는 값을 저장 해 둡니다.

print(c)
# c라는 이름이 바라보는 메모리의 값을 가지고 와서 
# 화면에 출력하는 기능 (print)을 수행

# - 기능 : 함수 (finction, method)라고 칭합니다.
#        : 이름뒤에 소괄호가 있으면 기능(함수)라고 생각하시면 됩니다.
#        : 자주 사용되는 처리 프로그램을 미리 만들어서 편하게 사용하게 끔 제공함
```

# 산술 연산자(+-*/ // %)

```python
'''
<산술 연산자>
- 더하기 : +
        : 문자열과 문자열 합치기 가능
- 빼기 : -
- 곱하기 : *
        : 문자열 같은 값 여러개 만들 때도 사용
- 나누기 : /
- 나눈 결과의 몫 : //
- 나눈 결과의 나머지 : %
- 제곱승 : ^
'''
```

```python
aa = 10
bb = 4

result = aa / bb 
result
```

```python
result = aa // bb
result
```

```python
result = aa % bb
result
```

```python
result = aa^3
result
```
