```python
### 데이터 처리를 위한 라이브러리 불러들이기
import pandas as pd
```

### 시리즈(Series) 구조 생성하기

```python
data = ["강아지","고양이","호랑이","사자","원숭이"]
idx_val = ["0", "1", "2", "3", "4"]

### 시리즈 구조 생성 :
# - index 속성값에 따라 행인덱스 값을 정의하고,
# - 각 인덱스 값에 해당하는 data값을 매핑해서 열단위 데이터를 생성함
# - 메모리가 만들어지고 주소를 반환함
pd.Series(data, index=idx_val)
```

### 데이터프레임 생성하기

##### 딕셔너리 데이터를 이용해서 -> 데이터프레임 생성하기

```python
dict_data = {"동물" : ["강아지","고양이","호랑이"], "나이" : [6,2,5]}
dict_data
```

```python
pd.DataFrame(dict_data)
```

##### 리스트 데이터를 이용해서 -> 데이터프레임 생성하기

```python
### 2차원 리스트로 만들어야 합니다. (3행 2열 데이터)
list_data = [["강아지", 7],["고양이", 2],["호랑이", 5]]
list_data

# 데이터프레임 생성
pd.DataFrame(list_data, columns=["동물이름","나이"])
```

##### 깨끗한 데이터 프레임을 생성해서 -> 행 데이터 추가하기

```python
df = pd.DataFrame()
df


df["동물이름"] = ["강아지", "고양이", "호랑이"]
df["나이"] = [2,45,6]
df["리스트"] = [list_data, ["1","2","3"],list_data]
df
```

```python
df = pd.DataFrame()
df["동물"] = [["강아지", "고양이"]]

dict_data = {"동물" : ["강아지", "고양이", "호랑이"], "나이" : [1,5,6]}
df["dict"] = [dict_data]

df["tuple"] = [("홍길동", "김유신", "이순신")]
df
```

```python
"""
df 데이터프레임 내에
 - 각 컬럼에 들어있는 데이터 중에 딕셔너리 타입의 데이터에 대해서는 
 - key와 value의 값을 각각 추출하여 아래와 같이 출력하기
 - 예시 : "강아지의 나이는 1살입니다."
          "고양이의 나이는 5살입니다."
          "호랑이의 나이는 6살입니다."
"""
```

```python
### 컬럼명만 추출하는 명령어
# -  columns : ["동물", "dict", "tuple"] 모든 컬렁명을 리스트로 반환함
df.columns
```

```python
for i in range(0,len(df["dict"][0]["동물"]),1) :
    print(f'{df[df.columns[1]][0]["동물"][i]}의 나이는 {df[df.columns[1]][0]["나이"][i]}살입니다.')
```
