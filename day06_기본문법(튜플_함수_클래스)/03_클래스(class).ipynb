```python
"""
<클래스(class)>
 - 객체(object)라고 칭함(객체지향 프로그램에서 사용되는 개념임)
 
 - 지금까지 사용한 일반변수, 여러값 가지고 다니는 변수, 함수 등을
   모아서 관리하는 메모리 공간(주소를 가짐)
   
 - 특정한 관리를 하나의 메모리 공간에 모아서 사용할 때 사용
  -> 예를 들어 회원관리라고 한다면, 회원 한명 한명에 다른 정보를 관리하는 클래스 생성
    --> 회원관리 클래스 내에 여러 기능들과 데이터를 관리하는 변수들을 정의하고,
    --> 외부에서 해당 클래스를 생성(메모리 주소 생성)하여 필요한 기능과 변수들을 이용함

<클래스 정의 방법>
### 클래스 이름의 첫글자는 대문자로 시작해야합니다.(규칙)
#   (함수는 주로 소문자로 시작합니다.)
 class 클래스이름(MemberInfo) :
     변수 정의


     ### 함수의 매개변수에 무조건 self를 넣어야합니다.(규칙)
     # - 이때 self는 멤버를 의미합니다.(해당 클래스에 속해있다라는 의미)
     # - self가 붙은 함수를 멤버 함수라고 칭함(외부에서 접근 가능한 함수를 의미함)


     ### 생성자 함수는 생략가능
     # - 단, self를 제외한 받아올 매개변수가 없는 경우 및
     #       처리할 문장이 없는 경우에
     def 생성자함수(self, 매개변수, ...) :
         처리 문장

     def 일반함수(self, 매개변수, ...) :
         처리 문장
         

"""
```

```python
### 자동차 정보 관리 클래스 정의하기

class CarInfo :
    ### 1. 생성자 함수는 생략하고 사용하겠음

    ### 2. 생성자 함수 생성하기(매개변수 한개 받기)
    # - 생성자 함수는 __init__(self, p_car_list)
    def __init__(self, p_car_list) :
        ### 자동차 정보를 관리하는 리스트 변수
        # - 자동차 정보 등록시 생성한 car_list변수는 사용 안함(주석처리)
        self.car_list = p_car_list
        
    
    ### 자동차 정보 1건 등록하는 함수 정의하기
    # - 자동차 정보 : 자동차 제조사, 이름, 색상, 배기량, 단가
    def setInsert(self, **p_carInfo) :
        ### 자동차 정보를 관리할 리스트 변수
        #self.car_list = []

        ### 자동차 정보 1건 리스트에 추가하기
        self.car_list.append(p_carInfo)

    ### 자동차 전체 정보 조회하는 함수 생성하기
    def getCars(self) :
        return self.car_list
```

```python
### 클래스 사용을 위해 생성합니다.
# - 이름, 객체 생성이라고 합니다.
# - 이때 객체의 의미는 메모리를 의미합니다.(주소가 생성 됩니다.)
# - 생성할 때 마다 새로운 주소가 만들어집니다.
#print(CarInfo(), CarInfo(), CarInfo())



### 생성은 클래스의 주소를 받아서 사용함
# - 주소를 모르면 생성된 메모리에 접근 할 수 없습니다.
#carInfo = CarInfo()


### 회원관리 리스트 변수를 생성해서 클래스 생성시 넘겨주기
carInfo_list = []

### 클래스 생성시 넘겨주는값이 있는 경우...
# - 클래스 생성자 함수를 클래스 내에 정의해서 매개변수로 받아서 처리해야 합니다.
carInfo = CarInfo(carInfo_list)


### 자동차 정보 입력하기 : 자동차 제조사, 이름, 색상, 배기량, 단가 넘겨주기
carInfo.setInsert(comp = "현대", name = "쏘나타", color = "흰색", cc = "2000cc", pay = 2500 )
```

### 생성자가 없는 클래스 정의하기

```python
### 더하기 연산이 포함된 Math 클래스 정의하기
# - 클래서 생성자 없음
class Math :
    ### 더하기 연산 기능
    def getPlus(self, p_num1, p_num2):
        return p_num1 + p_num2
        
```

```python
### 클래스 사용하기
# - 생성하기 : 클래스 생성자 함수가 자동으로 호출됩니다.
# -          : 클래스 생성시에는 무조건 생성자함수가 호출됨
#            : 클래스 정의 시에 생성자 함수가 없는 경우에는
#              디폴트(default) 생성자가 무조건 자동 호출됨
#              def __init__(self):
#            : 디폴트 생성자는 처리하는 문장이 없기 때문에 생략만 되어 있을 뿐
#              내부적으로 있다고 생각하시면 됩니다.

### 클래스 생성하기
# - 생성하게 되면 디폴트 생성자에 의해서
#   -> 메모리가 만들어지고 주소가 생성됩니다.
#   -> 이 주소를 반환받아서 사용하게 됨
# - 주소를 받은 math 변수를 클래스변수 또는 객체라고 칭합니다.
math = Math()
math
```

```python
### 클래스 내부 기능 사용하기
# - math 주소 안으로 접근(.)하여, 접근가능한 함수 호출하여 사여
# - 점(.)의 의미 : 메모리로 접근하여 "접근 지정자"라고 칭합니다.
# - 점(.)을 통해 접근 가능한 변수 또는 함수를 "멤버변수" 또는 "멤버함수"라고 칭합니다.
math.getPlus(3, 7)
```

### 생성자가 있는 클래스 정의하기

```python
class Math:
    ### 생성자 함수 정의하기
    # - 클래스 생성시 미리 멤버 변수에 값을 "정의"하거나,
    #                      멤버 함수를 호출하여 "처리"하고자 할 때 사용
    # - 생략하면 디폴트 생성자 함수가 자동 호출됨
    # - 생성자 함수는 일반 함수와 동일하게 정의하며
    #   --> 단, 함수 이름은 __init__()으로 사용해야함(규칙)
    # - 생성자 함수는 -> 클래스 생성시 최초 1회만 호출됨
    def __init__(self, a, b) :
        ### 함수 정의시 사용한 a, b변수는 지역변수(함수 내에서만 접근 가능)라고 하며,
        # - self가 붙은 aa, bb변수는 전역변수(클래스 내부 및 외부에서 접근 가능)라고 칭함
        self.aa = a
        self.bb = b
    ### 일반 함수 정의하기
    def getPlus(self):
        return self.aa + self.bb
        
```

```python
### 클래스 생성하기
# - 클래스 내에 생성자 함수를 정의하여 매개변수 2개의 값을 받게끔 해 놓았음
# - 따라서, 생성시 괄호안에 2개의 인자값을 넣어 주어야 합니다.
math = Math(30, 40)
math
```

```python
### 생성된 객체 주소를 통해서 클래스 내부로 접근하기
# - 멤버 변수 또는 멤버 함수들에 접근 가능
math.aa , math.bb, math.getPlus()
```
