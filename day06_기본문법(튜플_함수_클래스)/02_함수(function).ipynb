```python
### 함수로 생성하기
# 데이터의 정의 부분은 함수에 사용하지 않음
#  - 함수에서는 값을 받아서 처리
#  - 결과는 반환(return)하는 방식을 주로 사용
# <이름 규칙(정해지지는 않으나, 회사에서 주로 사용하는 규칙을 따름)>
# - 처리 결과를 반환 받을 때는 getXxx()로 주로 정의
# - 처리 결과를 저장 할 때는 setXxx()로 주로 정의
# - 조회 할 때는 getSelect() or getSearch() 등등...
# - 저장 할 때는 setInsert()
# - 수정 할 때는 setUpdate()
# - 삭제 할 때는 setDelete()
# - 처리 결과를 시각화 할 때는 viewXxx() or showXxx()로 주로 정의

### 매개변수 : 어떤 값이 넘어올지 모르는값을 받기 위한 임의 변수(없을 수도 있음)
# - return : 처리 결과를 호출한 쪽으로 반환할 때 사용(사용하지 않아도 됨)
# - 함수는 실행 시 문법적 오탈자 이외에 처리 오류는 확인하기 어려움
#   : 호출 시에 결과값을 확인하여 정확성 확인해야함
# for 매개변수 in range(0, 10, 1):
# *** 함수 정의
def getPlus(a_parameter, b_parameter) : 
    total = a_parameter + b_parameter
    # return은 문장의 제일 마지막에 사용합니다.
    return total
    

```

```python
### 함수로 연산 처리 문장 부분을 정의하기
# 함수명 : getMath()
def getMath(p_mode="+", p_num1=0, p_num2=0) :
    rtn_total = 0
    
    # 더하기 연산 조건 처리
    if p_mode == "+" :
        rtn_total = p_num1 + p_num2
    # 빼기 연산 조건 처리
    elif p_mode == "-":
        rtn_total = p_num1 - p_num2
    # 곱하기 연산 조건 처리
    elif p_mode == "*":
        rtn_total = p_num1 * p_num2
    # 나누기 연산 조건 처리
    elif p_mode == "/":
        # # 나누기는 분모가 0일 수 없으므로 조건 처리
        # if p_num2 > 0 :
        #     rtn_total = p_num1 / p_num2
        # else :
        #     print("분모는 0일 수 없습니다. 두번째 값을 다시 입력하세요!")
        rtn_total = p_num1 / p_num2
        

    ### 최종 결과값 반환(return)하기
    return rtn_total
```

```python
### 여러개의 값을 하나의 매개변수로 전달 받으려면?
# - 매개변수명 왼쪽에 *표시를 붙임
#   -> 타입은 튜플() 타입으로 받게 됨
def getTotalSum(*p_nums) :
    #print(p_nums)
    ### 넘겨 받은 숫자의 범위 값을 모두 더하기
    total = 0
    for num in p_nums : 
        total = total + num
    ### 모두 더한 값을 반환하기
    return total
```

```python
dict_list = {}
list1 = []
close = 0

# 신규입력 함수 정의
def setInsert(list1, name) :
    for i in list1:
        if i["name"] == name:
            print("이미 존재하는 이름입니다.")
            break
            
    else:
        age = input("나이를 입력해주세요")
        area = input("지역를 입력해주세요")
        list1.append({"name": name,"age": age,"area": area})
        print("정상적으로 추가되었습니다.")


# 수정하기 함수 정의

def setUpdate (list1, change_info):
    
    if change_info == "이름" :
        w_name = input("잘못 입력한 이름 입력")
        for i in range(0,len(list1),1):
            if list1[i]["name"] == w_name :
                change_name = input("다시 입력할 이름")
                list1[i]["name"] = change_name
                print(f"{w_name}님에서 {change_name}님으로 변경되었습니다.")
                return change_name
                break
                
    if change_info == "나이" :
        w_age = input("잘못 입력한 나이 입력")
        for i in range(0,len(list1),1):
            if list1[i]["age"] == w_age :
                change_age = input("다시 입력할 나이")
                list1[i]["age"] = change_age
                print(f"{w_age}살에서 {change_age}살로 변경되었습니다.")
                return change_age
                break
        
    if change_info == "지역" :
        w_area = input("잘못 입력한 지역 입력")
        for i in range(0,len(list1),1):
            if list1[i]["area"] == w_area :
                change_area = input("다시 입력할 지역")
                list1[i]["area"] = change_area
                print(f"{w_area}지역에서 {change_area}지역으로 변경되었습니다.")
                return change_area
                break

                
# 삭제하기 함수 정의
def setDelete (list1, del_info) :
    
    for i in range(0, len(list1), 1):
        if list1[i]["name"] == del_info:
            chk = input(f"{del_info} 삭제하시겠습니까? (예)")
            if chk == "예":
                del list1[i]
                print("삭제완료")
            break
    else:
        print("해당정보가 없습니다.")



# 조회하기 함수 정의
def getSelect (list1, sh_name) :
    for i in list1 :
        if i["name"] == sh_name :
            print(f'이름 {i["name"]} 나이 {i["age"]} 지역 {i["area"]}')
            break
    else :
        print("해당정보가 없습니다.")

        
#=====================================================================================================================


while 1:

    if close == "종료":
        
        break
    
    chlist = input("아래 번호 중에 하나를 선택해주세요.1. 리스트에 신규값 추가하기 2. 수정하기 3. 삭제하기 4. 검색(이름으로 조회) 5. 검색(전체 목록 조회) 6. 종료")


    if chlist == "1" :
        
        name = input("이름을 입력해주세요")
        setInsert(list1, name)
        
    if chlist == "2" :
        
        change_info = input("수정하실 정보를 입력(이름,나이,지역)")
        setUpdate(list1, change_info)
        
    if chlist == "3" :
        
        del_info = input("삭제할 정보의 이름을 입력해주세요")
        setDelete(list1, del_info)

    if chlist == "4" :
        
        sh_name = input("조회하실 정보의 이름을 입력해주세요")
        getSelect(list1, sh_name)
            
    if chlist == "5" :
        # 검색(전체목록 조회)
        for i in range(0,len(list1),1) :
            print(f"이름 : {list1[i]['name']} 나이 : {list1[i]['age']} 지역 : {list1[i]['area']}" )
    
    if chlist == "6" :
        close = "종료"
```

```python
# 빈 리스트를 만든다
list_name = []
list_addr = []
list_age = []
close = 0
# 신규 입력 함수 정의
def setInsert2 (list_name, name) :
    if name not in list_name :
        list_name.append(name)
        addr = input("주소을 입력해주세요")
        list_addr.append(addr)
        age = input("나이을 입력해주세요")
        list_age.append(age)
    
    else:
        print("중복된 이름입니다 다시 입력해주세요")

# 수정하기 함수 정의
def setUpdate2(list_name, change) :
    while 1 :
        
        change_infor = input("수정할 정보를 입력해주세요(이름, 주소, 나이)")
        if change_infor == "이름" :
            break 
        elif change_infor == "주소" :
            break
        elif change_infor == "나이" : 
            break

        if change_infor not in change :
            print("올바른 입력이 아닙니다. 다시 입력해주세요.")


    if change_infor == "이름":
        change_name = input("수정하고싶은 이름을 입력해주세요.")
        for i in range(0,len(list_name),1):
            if change_name == list_name[i]:
                change_name1 =input("무엇으로 변경해드릴까요?")
                list_name[i] =change_name1
                print(f"{list_name[i]}로 변경되었습니다")
                break
            else :
                print("이름이 틀립니다. 다시 시작해주세요")
    if change_infor == "주소":
        change_addr = input("수정하고싶은 주소을 입력해주세요.")
        for i in range(0,len(list_addr),1):
            if change_addr == list_addr[i]:
                change_addr1 =input("무엇으로 변경해드릴까요?")
                list_addr[i] =change_addr1
                print(f"{list_addr[i]}(으)로 변경되었습니다")
                break
            else :
                print("주소가 틀립니다. 다시 시작해주세요.")
    if change_infor == "나이":
        change_age = input("수정하고싶은 나이을 입력해주세요.")
        for i in range(0,len(list_age),1):
            if change_age == list_age[i]:
                change_age1 =input("무엇으로 변경해드릴까요?")
                list_age[i] =change_age1
                print(f"{list_age[i]}세로 변경되었습니다")
                break
            else :
                print("나이가 틀립니다. 다시 시작해주세요.")

# 삭제하기 함수 정의
def setDelete2(list_name, remover) :
    
    remove1 = list_name.index(remover)
    agree = input(f"{list_name[remove1]}님의 정보 이름:{list_name[remove1]}, 주소:{list_addr[remove1]}, 나이:{list_age[remove1]}를 정말 삭제하시겠습니까? 맞으시면 '예' 아니면 '아니요'를 쳐주세요")
    if agree == "예" :
        for i in range(0, len(list_name), 1):
            if remover == list_name[i]:
                del list_name[remove1]
                print("처리 완료")
                break

# 이름으로 조회하기 함수 정의
def getSearch2(list_name, sh_name) :
    try:
        sh_name2 = list_name.index(sh_name)
        print(f"{list_name[sh_name2]}님의 정보 이름:{list_name[sh_name2]} 주소:{list_addr[sh_name2]} 나이:{list_age[sh_name2]}")
    except :
        print(f"{sh_name}이 없습니다")

def getSelectAll (list_name):
    for i in range(0, len(list_name), 1) :
        print(f"{list_name[i]}님의 정보 이름:{list_name[i]} 주소:{list_addr[i]} 나이:{list_age[i]}")

#===============================================================================================================




while 1: 

    if close == "종료" :
        break
    
    chlist = input("아래 번호 중에 하나를 선택해주세요.1. 리스트에 신규값 추가하기 2. 수정하기 3. 삭제하기 4. 검색(이름으로 조회) 5. 검색(전체 목록 조회) 6. 종료")
    
    if chlist == "1" :   
        name = input("이름을 입력해주세요")  
        setInsert2(list_name, name)

    if chlist == "2" :
        change = ["이름","주소","나이"]
        setUpdate2(list_name, change)
        
    if chlist == "3" :
        remover = input("삭제할 정보를 입력해주세요")
        setDelete2(list_name, remover)

    if chlist == "4" :
        sh_name = input("조회하실 이름을 입력해주세요")
        getSearch2(list_name, sh_name)
       
    if chlist == "5" :
        getSelectAll (list_name)

    # 6을 입력받으면
    if chlist == "6" :
        # close 에 종료를 넣는다
        close = "종료"
    
```

```python
### 사용자로부터 입력을 받기 위한 기능으로 사용할 함수 정의
def getInput(mode, msg) :
    rtn = ""
    ### 문자열로 입력 받을 때 사용
    if mode == "str" :
        rtn = input(msg)
        
    ### 정수값으로 입력 받을 때 사용
    elif mode == "int" :
        try :
            rtn = int(input(msg))
        except :
            print("숫자만 입력 가능합니다.")
            rtn = "error"
            
    return rtn
    
### 중복된 이름 체크하는 함수
def duplicationCheck(p_list, p_name) :
    if p_name in p_list :
        print("중복된 이름이 있습니다. 다시 입력하세요!")
        return 1
    else :
        return 0

### 수정할 대상 이름이 있는지 체크하는 함수 
def nameCheck(p_list, p_name):
    if p_name not in p_list :
        print("수정할 대상 이름이 존재하지 않습니다. 다시 입력하세요!")
        return 1
    else : 
        return 0

### 이름 중복 또는 존재여부 확인하는 함수를 통합하기
def nameCheckAll(mode, p_list, p_name):
    ### 중복 확인 == 1
    if mode == 1 :
        return duplicationCheck(p_list, p_name)
    ### 존재여부 확인 == 2
    else :
        return nameCheck(p_list, p_name)

### 최초에 사용자에게 보여질 메시지를 변수에 저장해서 while문 안에서 input(msg)함수에 사용
main_msg = """아래 번호 중에 하나를 선택해 주세요(번호로 입력하여 주세요) :
       1. 리스트에 신규값 추가하기
       2. 수정하기
       3. 삭제하기
       4. 검색(이름으로 조회)
       5. 검색(전체 목록 조회)
       6. 종료
"""

### 회원정보 담다놓을 리스트 변수
member_list = []

while 1 :
    ### 입력 받기
    num = getInput("int", main_msg)

    ### 입력 받은 값이 정수가 아닌경우 다시받기
    if num == "error" : continue

    if num == 1 :
        print("입력----------------")
        name = getInput("str", "이름을 입력하세요 : ")

        ### 중복된 이름 체크하기
        # if duplicationCheck(member_list, name) : continue
        if nameCheckAll(1, member_list, name) : continue
        
        age  = getInput("int", "나이를 입력하세요 : ")
        if age == "error" :  continue
        
        area = getInput("str", "지역을 입력하세요 : ")
        print(name, age, area)

    elif num == 2 :        
        print("수정----------------")
        name = getInput("str", "수정할 대상의 이름을 입력하세요 : ")

        ### 수정할 대상 이름이 있는지 체크하기
        if nameCheckAll(2, member_list, name) : continue
        
    elif num == 6 :
        print("프로그램을 종료 합니다.")
        break
        

```
