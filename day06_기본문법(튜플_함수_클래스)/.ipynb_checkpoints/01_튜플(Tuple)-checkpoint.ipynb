```python
"""

<튜플(tuple)>
 - 리스트(list)와 동일한 메모리 구조를 가지고 있음
 - 따라서, 리스트 사용법과 동일하게 사용가능(인덱스 번호를 가지고 있음)
 - 리스트와의 차이점
   -> 소괄호() 기호를 사용함
   -> 수정을 할 수 없음 (수정을 하기 위해서는 삭제 후 다시 생성해야 함)
 - 튜플은 주로 정의해서 사용하기 보다는
   -> 임의 처리 결과를 받아서 사용하는 경우가 많음
   -> 예를 들어 zip() 함수의 결과 : [(값1, 값2), (값3, 값4)...]
   ->          여러 변수의 값 출력 결과 : a, b -> (a값, b값)

"""
```

```python
a = 10
b = 20
c = 30


a, b, c
```

```python
### d변수의 타입 : 튜플 타입
d = a, b, c
d
```

```python
### 리스트와 동일하게 인덱스 번호로 접근가능
for idx in range(0, len(d), 1):
    print(d[idx])
```

```python
### 리스트와 동일하게 튜플 자체값을 사용할 수 있음
for v in d :
    print(v)
```

```python
### 튜플은 값의 수정이 불가능 합니다.
d[1] = 30
```

```python
a = [10,20,30]
b = ["홍", "김", "이"]
c = ["서울", "대전", "부산"]

list(zip (a, b, c))
```
