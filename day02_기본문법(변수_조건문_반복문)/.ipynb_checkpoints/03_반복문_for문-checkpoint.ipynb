```python
### 1부터 10까지의 순차적인 총합을 출력하기(for문 이용)
# - 패턴 찾기 : 1씩증가하는 패턴이 있음
# - 반복 범위 : 시작값=1, 끝값=10

### 주로 기본적으로 사용되는 for문의 형태
# - range(시작값, 끝값, 증가 패턴) : 범위 생성 함수(기능)
# - range(1, 10, 1)의 해석
#   : 1부터/ 10보다 1작을 때까지/ 1씩 증가하면서 순차적인 값을 생성
#   : [1, 2, 3, 4, 5, 6, 7, 8, 9]라는 형태의 값을 만들어 냅니다.
#   : 생성된 9개의 값을 왼쪽ㅇ no라는 변수에 한개씩 넣어서 반복시켜 줍니다.

result = 0
for no in range(1,11,1): 
    #print(f"{no} 번째 반복 중...")
    result = result + no 
    print(f"{no}번째 누적값 result={result}")
print("for문이 종료됩니다.")
```

```python
### 1부터 100까지의 총합을 출력해 주세요

result = 0
#for i in range(1,101,1):
#for i in range(1,101):
# range(기본값=0 끝값, 기본값=1) 
for i in range(101):
    result = result + i
print(f"1부터 100까지의 총합은 {result}입니다.")
```

```python
### 1부터 100까지의 정수값 중에
# 짝수값만 출력해주세요...

for i in range (1, 101, 1):
    # 짝수인 경우에만 출력하기(조건 처리)
    if i % 2 == 0:
        print(i)
```

```python
### 1부터 100까지의 정수값 중에
# 짝수값의 총합만 출력해주세요...
### 전역변수 : 어디서든지 사용되는 변수
# 지역변수 : 특정 블록 내에서 정의된 변수(블럭 내에서만 사용) *jupyter notebook은 지역변수도 밖에서도 쓸 수 있다. 보통은 안됨
result = 0
for i in range(1,101,1):
    if i % 2 == 0 :
        #result = result + i
        result  += i
print(f"1부터 100까지의 짝수값의 총합은 {result}입니다")
```

```python
### 임의 정수값 2개를 입력 받기
# - 두번째 값은 첫번째 값보다 큰 값을 입력 받기
# - 첫번째 값부터 두번째 값까지의 총합을 출력해 주세요

result = 0
a = int(input("첫번째 값을 입력해주세요"))
b = int(input("두번째 값을 입력해주세요(단, 첫번째 값보다 큰 값을 입력해주세요)"))

### 두번째 값은 첫번째 값보다 큰값을 입력 받기
if a < b :
    ### for문의 내부 반복 조건
    # - 시작값 < 끝깞 이라는 조건을 만족해야 반복 가능
    for i in range(a, b+1, 1):
        result += i
    print(f"첫번째 값부터 두번째 값까지의 총합은{result}입니다.")

else : 
    print("두번째 값은 첫번째 값보다 커야합니다. 다시 입력해 주세요.")
```

```python
### 2024년 1월부터 2025년 12월까지 아래와 같이 출력해주세요
# 출력 예시
#  "2024년1월입니다"
#  "2024년2월입니다"
#         ...
#  "2025년12월입니다"

for i in range (2024, 2026, 1):
    for j in range (1, 13, 1):
        result += j
        print(f"{i}년 {j}월입니다.")
```

```python
### 2024년 1월 1일부터 2025년 12월 31일까지의 값을 출력
# - 월은 1~12월까지
# - 일은 1~31일까지 있다는 것으로 가정합니다.
# 출력예시 : "2024-1-1"
#                 ....
#            "2025-12-31"
for i in range (2024, 2026, 1):
    for j in range (1, 13, 1):
        for u in range (1, 31 ,1):
            result += u    
            print(f"{i}-{j}-{u}")
```

```python
### 2024년 1월 1일부터 2025년 12월 31일까지의 값을 출력
# - 월은 1~12월까지
# - 일은 1~31일까지 있다는 것으로 가정합니다.
# 출력예시 : "2024-01-01"
#                 ....
#            "2025-12-31"
for i in range (2024, 2026, 1):
    for j in range (1, 13, 1):
        if len(str(j)) == 1:
            j = "0" + str(j)
        for u in range (1, 31 ,1):
            if len(str(u)) == 1:
                u = "0" + str(u)               
            print(f"{i}-{j}-{u}")
```

```python
### 문자열 자릿수 정의 함수 : zfill()
# - zfill() : 문자열 함수, 
# - 작성 방법
#   "문자".zfill(자릿수) -> "3".zfill(2) => "03" 
#                        -> '3'.zfill(3) => "003"
#                        -> '33'.zfill(2) => "33"
```

```python
### 2024년 1월 1일부터 2025년 12월 31일까지의 값을 출력
# - 월은 1~12월까지
# - 일은 1~31일까지 있다는 것으로 가정합니다.
# 출력예시 : "2024-01-01"
#                 ....
#            "2025-12-31"
for i in range (2024, 2026, 1):
    for j in range (1, 13, 1):
        for u in range (1, 31 ,1):    
            print(f"{i}-{str(j).zfill(2)}-{str(u).zfill(2)}")
```
