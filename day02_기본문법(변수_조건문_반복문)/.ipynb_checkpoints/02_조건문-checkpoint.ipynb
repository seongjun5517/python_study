```python
'''

<조건문>
 - 2개 이상의 값을 비교하고자 할 때 사용
 - 보통 조건 연산자(비교 연산자)를 이용하여 사용
 - 조건(비교) 연산자 : 크다, 작다, 크거나 같다, 작거나 같다, 같다, 같지않다
                     :  >     <         >=          <=       ==      !=
 
 - 규칙 : 조건문은 if로 시작해야함
        : 조건 뒤에는 세미콜론(:) 넣어야 합니다.(처리 프로그램의 블록을의미함)
        : 세미콜론(:) 밑에는 항상 들여쓰기로 처리 프로그램을 작성해야함
 - 작성방법 : if 값1 > 값2 :
                처리프로그램
              elif 값1 > 값 3 : 
                처리프로그램
              elif ... :
                ...
              else : 
                        위 조건이 모두 만족하지 않으면 처리
 - if ~ elif ~ else : if만 작성해도 되며, if ~ else만 작성해도 됨
'''
```

```python
### 조건을 이용한 처리 프로그램 작성 : 조건문이라고 칭함
a = 10
b = 3

### (시나리오)
# a의 값이 b의 값보다 크면 -> "a가 b보다 큽니다"를 출력하기

# 1. 조건 만들기(조건에 따라 달라짐)
if a > b :
    # 2. 처리 문장 작성(시나리오에 따라 달라짐)
    print("a가 b보다 큽니다")
    
```

```python
### (시나리오)
# a의 값이 b의 값보다 크면 -> "a가 b보다 큽니다"를 출력하고,
# a의 값이 b의 값보다 작으면 -> "a가 b보다 작습니다."를 출력하기
# -> a의 값을 수정하면서 테스트하기

a = 1

# 1. 조건 만들기(조건에 따라 달라짐)
if a > b :
    # 2. 처리 문장 작성(시나리오에 따라 달라짐)
    print("a가 b보다 큽니다")
# 3. 다른 조건 만들기
elif a < b :
    # 4. 처리 문장 작성
    print("a가 b보다 작습니다.")
```

### 조건문 안에 조건문 넣기 : 중첩(다중) 조건문 이라고도 합니다.

```python
### 회원 중에 홍길동이라는 회원이 들어오면,
# - "홍길동 회원님 환영합니다"
# - 그리고 홍길동 회원의 나이가 30대 이상인지, 아닌지도 함께 출력

# - 출력 예시
#   "홍길동 회원님 환영합니다."
#   "나이는 30대 이상입니다. " or "30대가 아닙니다."
# - 1조건 : 홍길동 인지 아닌지
# - 2조건 : 홍길동 이라면 -> 30대 이상인지 아닌지

name = "홍길동"
age = 30

if name == "홍길동" :
    print("홍길동 회원님 환영합니다.") 
    if age >= 30 :
        print("30대 이상입니다.")
    else : 
        print("30대 이상이 아닙니다.")
else :
    print("홍길동 회원님이 아닙니다.")
    
    
```

### 논리 연산자

```python
'''
<논리 연산자>
 - 2개 이상의 조건을 처리 동시에 처리 할 때 사용함
 - 조건문에 조건과 조건 사이에 사용됩니다.
 - 논리연산자 : and(모든 조건이  참인 경우), or (조건 중에 하나라도 참인 경)
 - 사용방법
    if (값1 > 값2) and (값3 == 값4) :
        처리문장 작성...
        
    if (값1 > 값2) or (값3 == 값4) :
        처리문장 작성...
        

'''
```

```python
### 시나리오
# - hong이라는 아이디와, asdf라는 패스워드를 가지는 회원에 대해서는 
# - 아이디와 패스워드가 모드 일치 합니다. 라고 출력
# - 하나라도 맞지않으면, "다시 입력해주세요" 출력


id = "hong"
password = "asdf"

if (id == "hong") and (password == "asdf") : 
    print("모두 일치합니다.")
else :
    print("다시 입력해주세요.")
```

### 값 입력 받기(input())

```python
'''
<입력 함수 (기능)>
 - input()
 - 사용방법
   변수 = input("값을 입력해주세요 : ")
 - input() 함수가 입력된 값을 반환할 때 문자열타입으로 반환해 줍니다.
 - 입력은 키보드로 입력을 하면, input()함수가 알아서 문자열로 반환해 줍니다.
 
'''
```

```python
### 사용자로 부터 이름을 입력 받기 : input("이름을 입력해 주세요 :")
# - 입력 받은 이름이 "홍길동" 이면
# -> "홍길동님 안녕하세요" 를 출력
# - "홍길동"이 아니면
# -> "넌 누구니?"를 출력
# 단, 출력할 때 이름은 입력 받은 변수를 사용.

### 사용자 이름 입력 받기
name = input("이름을 입력해주세요.")

### 조건처리
if name == "홍길동" :
    print(f"{name}님 안녕하세요.")
else :
    print(f"{name} 너 누구니?")
    
```

```python
'''
나이를 입력 받아서
 - 나이가 20세 미만이면 "청소년 할인" 을 출력
 -        20세 이상이면서 60세 미만이면 "일반인" 을 출력
 -        60세 이상이면 "실버할인" 을 출력
 -        위 조건을 모두 만족하지 않으면 무조건 "알수없음" 을 출력
 
'''
age = int(input("나이를 입력해주세요."))

if age < 20 :
    print("청소년 할인")
elif (age >= 20) and (age < 60) :
    print("일반인")
elif age >= 60 :
    print("실버할인")
else :
    print("알수없음")
```

```python
"""
과목 1개에 대한 점수를 입력받아서
 - 입력 받은 점수에 따른 학점을 출력하고자 합니다.
 * 100~95점은 "A+" 출력
   94~90점은 "A0" 출력
   89~85점은 "B+" 출력
   84~80점은 "B0" 출력
   79점 이하 점수는 "F" 출력
"""


grade = int(input("점수를 입력해주세요."))

if grade >= 95 :
    print("A+")
elif grade >= 90 :
    print("A")
elif grade >= 85 :
    print("B+")
elif grade >= 80 :
    print("B")
else :
    print("F")
```

```python
### 정수값 입력 받기
in_data = int(input("임의 정수값을 입력하세요."))
#in_data


## 입력받은 값이 짝수인지 홀수인지 판별하기 : 조건문 사용하기
# - 짝수 및 홀수에 대한 계산은 산술 연산자 % 사용 : 나눈 나머지의깂이 0 or 1

if (in_data % 2) == 0 :
    print(f"입력받은 값은 {in_data}이며, 짝수입니다.")
else :
    print(f"입력받은 값은 {in_data}이며, 홀수입니다.")
```

```python
### 값 3개를 입력받기
# - 첫번째 값 : +-*/의 사칙연산 기호를 입력받기
# - 두번째 값 : 임의 정수값 입력 받기
# - 세번째 값 : 임의 정수값 입력 받기
# ==> 입력 받은 두번째와 세번째 값을 이용해서 
#     첫번째 값의 연산 기호에 따라 연산한 결과값을 출력하기
# ==> print()는 한번만 사용
# - 최종 결과 값 예시 : +, 1, 2를 입력 받은 경우 " 1 + 2 + = 3입니다." 처럼 출력


clac = input("사칙연산 기호를 입력해주세요(+, -, *, /)")
a = int(input("첫번째 임의 정수값을 입력해주세요."))
b = int(input("두번째 임의 정수값을 입력해주세요."))
c = 0

if clac == '+' :
    c = a + b
elif clac == '-' :
    c = a - b
elif clac == '*' :
    c = a * b
elif clac == '/' :
    if b == 0 :
        print("분모의 값은 0이 될 수 없습니다. 다시 입력해 주세요")
    else :
        c = a / b
else :
    print("예의상 처리 : 잘못 입력하셨네요. 다시 입력해 주세요")
print(f"{a}{clac}{b}={c}입니다.")



### 사칙연산도 문자열로 처리한다    
```
